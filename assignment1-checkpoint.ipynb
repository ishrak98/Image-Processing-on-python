```python
import numpy as np
import cv2
path1=r"C:\Users\ayon5\OneDrive\Desktop\image processing project\book images\front1.jpg"
path2=r"C:\Users\ayon5\OneDrive\Desktop\image processing project\book images\front2.jpg"
img1=cv2.imread(path1,1)
img2=cv2.imread(path2,1)
gray1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gray2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
sift=cv2.xfeatures2d.SIFT_create()
kp1,des1=sift.detectAndCompute(gray1,None)
kp2,des2=sift.detectAndCompute(gray2,None)
bf=cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
p1=(des1*255).astype(np.uint8)
p2=(des2*255).astype(np.uint8)
matches=bf.match(p1,p2)
matches=sorted(matches,key= lambda x:x.distance)
matching_result=cv2.drawMatches(img1,kp1,img2,kp2,matches[:10],None,flags=2)
cv2.imshow("image",matching_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
